# Azure AI Translator Portal
Welcome to the Azure AI Translator Portal. Choose one of the options below to get started.

```python
import ipywidgets as widgets
from IPython.display import display, HTML
```

```python
# Create buttons for each functionality
text_translation_button = widgets.Button(description="Text Translation", button_style='info')
document_translation_button = widgets.Button(description="Document Translation", button_style='info')
voice_translation_button = widgets.Button(description="Voice Translation", button_style='info')
```

```python
# Define button click event handlers
def on_text_translation_button_clicked(b):
    display(HTML('<a href="4 Translation.ipynb" target="_blank">Go to Text Translation</a>'))

def on_document_translation_button_clicked(b):
    display(HTML('<a href="5 Document translation batch.ipynb" target="_blank">Go to Document Translation</a>'))

def on_voice_translation_button_clicked(b):
    display(HTML('<a href="7 Vocal Translator using Azure AI.ipynb" target="_blank">Go to Voice Translation</a>'))
```

```python
# Assign event handlers to buttons
text_translation_button.on_click(on_text_translation_button_clicked)
document_translation_button.on_click(on_document_translation_button_clicked)
voice_translation_button.on_click(on_voice_translation_button_clicked)
```

```python
# Display buttons
display(widgets.HBox([text_translation_button, document_translation_button, voice_translation_button]))
```
